### Relationship between Backpropagation and Gradient Descent:

- **Backpropagation** is the method used to calculate **rate of change (Gradients)**, determining how each weight in the neural network affects the error.  
- **Gradient Descent** utilizes these **rates** to update the weights and gradually reduce the error.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y,iterations,lr):
    # assume random values for m and b
    m = 0
    b = 0
    n = len(x)
    for i in range(iterations):
        y_pred = m * x + b

        dedm = -2/n * sum((y - y_pred) * x)
        dedb = -2/n * sum(y - y_pred)
        
        m = m - lr * dedm
        b = b - lr * dedb

        print(sum(y - y_pred))
   
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y,1000,0.1)
```

```python
gradient_descent(x,y,1000,0.01)
```

```python
gradient_descent(x,y,1000,0.001)
```
