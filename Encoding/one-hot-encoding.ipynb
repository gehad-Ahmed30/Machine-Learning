## Using pandas to create dummy variables

```python
town_dummy=pd.get_dummies(data.town).astype(int)
town_dummy
```

```python
merged=pd.concat([data,town_dummy],axis='columns')
merged
```

```python
merged.drop(['town'],axis='columns',inplace=True)
```

```python
merged
```

```python
merged.drop(['Haram'],axis='columns',inplace=True)
merged
```

```python
x=merged.drop(['price'],axis='columns')
y=merged.price
```

```python
x
```

```python
y
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[2600,0,0]])
```

```python
from sklearn.metrics import  mean_absolute_error,mean_absolute_percentage_error
y_pred=model.predict(x)
```
