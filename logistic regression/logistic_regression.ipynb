```python
import math
def sigmoid(y):
    return 1/(1+math.exp(-y))
```

```python
def prediction_function(age):
    z=0.1094324 * age -3.92618657 # 0.04150133 ~ 0.042 and -1.52726963 ~ -1.53
    y=sigmoid(z)
    return y
```

```python
age=27
prediction_function(age)
```
